```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x * (1 - x)

# Generate 5 equally spaced points between 0 and 1
x_points = np.linspace(0.1, 0.8, 5)

# Evaluate the function at those points
y_points = f(x_points)

# Add Gaussian noise to the y-values (mean=0, std=0.05)
noise = np.random.normal(0, 0.01, size=y_points.shape)
y_points_noisy = y_points + noise

# Generate a smoother range of x values for plotting the function
x_values = np.linspace(0, 1, 400)
y_values = f(x_values)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=r'$m(x)$', color='blue')  # Function curve
plt.scatter(x_points, y_points_noisy, color='red', zorder=5, label='Points')  # Points

# Add dotted vertical lines from the points to the x-axis
for x, y in zip(x_points, y_points_noisy):
    plt.vlines(x, 0, y, color='gray', linestyles='dotted')


# Label the axes and the title
#plt.title('Plot of the function $f(x)$ and 5 equally spaced points')
plt.xlabel('x')
plt.ylabel('f(x)')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
