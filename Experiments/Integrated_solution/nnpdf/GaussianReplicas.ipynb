```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec


# Parameters for the normal distribution
mean = 0        # Mean (loc)
std_dev = 1     # Standard deviation (scale)
num_points = 10000  # Number of points to generate

# Generate normally distributed data
data = np.random.normal(loc=mean, scale=std_dev, size=num_points)

# Create a figure and axes for the subplots (1 row, 2 columns)
#fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Create a figure
fig = plt.figure(figsize=(14, 6))

# Create a GridSpec to control the size of subplots
gs = gridspec.GridSpec(1, 2, width_ratios=[2, 1])  # Subplot 1 twice as wide as subplot 2

# Create subplots using GridSpec
ax1 = fig.add_subplot(gs[0])  # Time series plot on the left (larger)
ax2 = fig.add_subplot(gs[1])  # Histogram plot on the right (smaller)

# Plot the time series (data points) on the first axis
ax1.plot(data, color='blue', label='Data')
ax1.set_xlabel('Replica')
ax1.set_ylabel('Value')
ax1.set_title('Time Series Plot for f(x)')
ax1.grid(True)

# Plot the histogram on the second axis
ax2.hist(data, bins=30, orientation='horizontal', color='skyblue', edgecolor='black', alpha=0.7)
ax2.set_xticks([])
ax2.set_yticklabels([])
ax2.set_xticklabels([])
#ax2.set_xlabel('Value')
#ax2.set_ylabel('Frequency')
#ax2.set_title('Histogram')
#ax2.grid(True)
for spine in ax2.spines.values():
    spine.set_visible(False)


# Adjust the layout to ensure proper spacing between subplots
fig.tight_layout()

# Show the plot
plt.show()
```
