```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data (x and y)
x = np.array([.1, .2, .3, .4, .5, .6, .7])
y = 2 + 3*x**2 + x**3

# Add Gaussian noise
mean = 0        # Mean of the noise
std_dev = 0.15  # Standard deviation of the noise
noise = np.random.normal(mean, std_dev, y.shape)  # Generate noise

# Add noise to the original data
y_noisy = y + noise

# Fit the noisy data to a polynomial of degree 1 (linear fit)
degree = 1
coefficients1 = np.polyfit(x, y_noisy, degree)

# Generate the polynomial equation
polynomial1 = np.poly1d(coefficients1)

# Print the polynomial
print("Polynomial equation: ", polynomial1)

# Fit the noisy data to a polynomial of degree 4 (quartic fit)
degree = 3
coefficients3 = np.polyfit(x, y_noisy, degree)

# Generate the polynomial equation
polynomial3 = np.poly1d(coefficients3)

# Print the polynomial
print("Polynomial equation: ", polynomial3)

# Fit the noisy data to a polynomial of degree 4 (quartic fit)
degree = 6
coefficients6 = np.polyfit(x, y_noisy, degree)

# Generate the polynomial equation
polynomial6 = np.poly1d(coefficients6)

# Print the polynomial
print("Polynomial equation: ", polynomial6)

# Generate values for plotting the fit line
x_fit = np.linspace(min(x)-0.07, max(x)+0.05, 100)
y_fit1 = polynomial1(x_fit)
y_fit3 = polynomial3(x_fit)
y_fit6 = polynomial6(x_fit)

# Plot the original data, noisy data, and the polynomial fit
plt.plot(x_fit,2 + 3*x_fit**2 + x_fit**3, color='black', linestyle='--')
#plt.scatter(x, y, color='green', label='Original Data', marker='x', s=100)
plt.scatter(x, y_noisy, color='red', label='Noisy Data', marker='o')
plt.plot(x_fit, y_fit1, label=f'Polynomial Fit (Degree 1)', color='blue')
plt.plot(x_fit, y_fit3, label=f'Polynomial Fit (Degree 3)', color='cyan')
plt.plot(x_fit, y_fit6, label=f'Polynomial Fit (Degree 6)', color='maroon')
plt.legend()
plt.ylim(1.5,4)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Fit with Gaussian Noise')
plt.show()
```
